## Visualization

```
import matplotlib.pyplot as plt
x=[3,4,5,6]
y=[1.5,2,2.5,3]
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```
x = np.arange(0,6,0.3)
plt.plot(x,x,'ro')
plt.plot(x,x**2,'bs')
plt.plot(x,x**3,'g')
plt.plot(x,x**4,':')
plt.xlim(0,6)
plt.ylim(0,125)
plt.show()
```

```
height = [179,155,191,152,188,177]
names = ['QA','WB','EC','RD','TE','YF']
plt.bar(names,height)
plt.title('Height Comparison')
plt.show()
```

```
x = np.random.randn(100000)
plt.hist(x,bins=10)
plt.show()
```
